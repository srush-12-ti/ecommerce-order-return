```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\srush\OneDrive\Desktop\datascience\product retun\order_dataset.csv" )

# Group by Buyer ID
refund_revenue_by_buyer = df.groupby('Buyer ID').agg({
    'Refunded Item Count': 'sum',
    'Final Revenue': 'sum'
}).reset_index()

print(refund_revenue_by_buyer.head())
```

```python
# If each row is an order:
total_orders = len(df)
print("Total Orders:", total_orders)
```

```python
total_orders = df['Category'].nunique()
print("Total Unique Orders:", total_orders)
```

```python
orders_by_buyer = df['Buyer ID'].value_counts()
print("Orders by buyer:")
print(orders_by_buyer.head(10))  # Top 10 customers

# Barplot for top 10 customers
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(x=orders_by_buyer.head(10).index, y=orders_by_buyer.head(10).values)
plt.title('Top 10 buyer by Number of Orders')
plt.xlabel('Buyer ID')
plt.ylabel('Number of Orders')
plt.show()
```

```python
orders_by_item = df['Item ID'].value_counts()
print("Orders by item:")
print(orders_by_item.head(10))  # Top 10 products

# Barplot for top 10 products
plt.figure(figsize=(12,6))
sns.barplot(x=orders_by_item.head(10).index, y=orders_by_item.head(10).values)
plt.title('Top 10 Products by Number of Orders')
plt.xlabel('Item ID')
plt.ylabel('Number of Orders')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation
corr = df['Total Revenue'].corr(df['Refunded Item Count'])
print(f"Correlation between Total Revenue & Refunded Item Count: {corr:.3f}")

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x='Refunded Item Count', y='Total Revenue', data=df)
plt.title('Total Revenue vs Refunded Item Count')
plt.xlabel('Refunded Item Count')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Calculate correlation
corr = df['Purchased Item Count'].corr(df['Price Reductions'])
print(f"Correlation between Purchased Item Count & Price Reductions: {corr:.3f}")

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x='Price Reductions', y='Purchased Item Count', data=df)
plt.title('Purchased Item Count vs Price Reductions')
plt.xlabel('Price Reductions')
plt.ylabel('Purchased Item Count')
plt.show()
```

```python
# Calculate correlation
corr = df['Final Quantity'].corr(df['Final Revenue'])
print(f"Correlation between Final Quantity & Final Revenue: {corr:.3f}")

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x='Final Quantity', y='Final Revenue', data=df)
plt.title('Final Quantity vs Final Revenue')
plt.xlabel('Final Quantity')
plt.ylabel('Final Revenue')
plt.show()
```
